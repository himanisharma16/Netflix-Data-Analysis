# Imputing and Handling Missing Values

```python
#number of null values in Dataset
df.isnull().sum()
```

```python
#imputing null values
df.director=df.director.fillna("unknown")
df.cast = df.cast.fillna('Not known')
df.country=df.country.fillna('unknown')
df.rating = df.rating.fillna("No Rating")
```

```python
# Dropping rows with null values
df.dropna(inplace=True)
```

```python
#again check the null values
df.isnull().sum()
```

```python
# checking the datatype of columns in the dadtset
df.dtypes
```

```python
#changing the datatype of"date_added" to datetime
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
#Null values arose due to inconsistent or missing data during type conversion
df.isnull().sum()
```

# Exploratory Data Analysis

```python
#Top 10 Countries Producing Netflix Content
sns.barplot(x=df.country.value_counts().head(10), 
            y=df.country.value_counts().head(10).index, 
            palette=["red", "black"] * 5)  
plt.show()
```

```python
plt.pie(df.country.value_counts().head(),labels=df.country.value_counts().head().index,startangle=90, autopct='%.1f%%')
plt.legend(loc='lower left', title="Country")
plt.title('Visualising content produced by top 5 countries')
plt.show()
```

```python
df.rating.value_counts().reset_index()
```

```python
sns.barplot(y=df.rating.value_counts().head(10), 
            x=df.rating.value_counts().head(10).index, 
            palette=['red', 'black'] * 5) 

plt.title('Rating Analysis', fontsize=14, fontweight='bold', color='black')
plt.xlabel('Ratings', fontsize=12, fontweight='bold', color='black')
plt.ylabel('Counts', fontsize=12, fontweight='bold', color='black')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='release_year', data=df, color='red')  # Set bar color to red

plt.title('Content Release Over the Years on Netflix', fontsize=14, color='black')
plt.xlabel('Release Year', fontsize=12, color='black')
plt.ylabel('Count', fontsize=12, color='black')
plt.xticks(rotation=90, color='black')
plt.yticks(color='black') 

plt.show()
```

```python
# Grouping data by year and type
content_trend = df.groupby(['release_year', 'type']).size().unstack()
```

```python
plt.figure(figsize=(12, 6))
content_trend.plot(kind='bar', stacked=True, color=['red', 'black'], figsize=(12, 6))

plt.title('Content Added Over the Years by Type', fontsize=14, color='black')
plt.xlabel('Release Year', fontsize=12, color='black')
plt.ylabel('Count', fontsize=12, color='black')
plt.legend(title="Type", labels=["Movie", "TV Show"])
plt.xticks(rotation=90, color='black')
plt.yticks(color='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['type'].value_counts().plot.pie(autopct='%1.1f%%',startangle=90, colors=['red', 'black'], textprops={'fontsize': 12, 'color': 'blue'}
)

plt.title('Distribution of Movies and TV Shows on Netflix', fontsize=14, fontweight='bold', color='black')
plt.ylabel('') 

plt.show()
```
